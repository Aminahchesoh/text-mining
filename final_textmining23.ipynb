# Normalization

```python
import re
import string

def normalize_text(word_list):
    # Join the list of words into a single string
    text = ' '.join(word_list)
    # ‡∏•‡∏ö‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏°‡∏≤‡∏¢‡∏ß‡∏£‡∏£‡∏Ñ‡∏ï‡∏≠‡∏ô
    text = re.sub(f"[{re.escape(string.punctuation)}]", '', text)
    # ‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡∏û‡∏¥‡∏°‡∏û‡πå‡πÄ‡∏•‡πá‡∏Å
    text = text.lower()
    return text

# Apply the modified normalize_text function to the 'remove' column
df['norm'] = df['remove'].apply(normalize_text)
df['norm']
```

# Stopword 

```python
from collections import Counter
from nltk.tokenize import word_tokenize 
from pythainlp.corpus.common import thai_stopwords  # ‡∏ô‡∏≥ stopwords ‡∏†‡∏≤‡∏©‡∏≤‡πÑ‡∏ó‡∏¢‡πÄ‡∏Ç‡πâ‡∏≤‡∏°‡∏≤

# ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô remove_stopwords
def remove_stopwords(text):
    # ‡πÅ‡∏ö‡πà‡∏á‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡∏≥
    words = word_tokenize(text)

    # ‡∏Å‡∏£‡∏≠‡∏á‡∏Ñ‡∏≥‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô stopwords
    filtered_words = [word for word in words if word not in thai_stopwords()]

    # ‡∏£‡∏ß‡∏°‡∏Ñ‡∏≥‡∏ó‡∏µ‡πà‡πÄ‡∏´‡∏•‡∏∑‡∏≠‡πÄ‡∏Ç‡πâ‡∏≤‡∏™‡∏π‡πà‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡πÉ‡∏´‡∏°‡πà
    filtered_text = ' '.join(filtered_words)

    return filtered_text

# ‡∏ñ‡πâ‡∏≤ 'clean_text' ‡πÉ‡∏ô DataFrame ‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ó‡∏µ‡πà‡∏Ñ‡∏±‡πà‡∏ô‡∏î‡πâ‡∏ß‡∏¢‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á
# ‡πÉ‡∏´‡πâ‡∏•‡∏ö stopwords
df['norm'] = df['norm'].apply(remove_stopwords)

# ‡∏ï‡∏±‡∏î‡∏Ñ‡∏≥‡πÉ‡∏ô‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏•‡∏ö stopwords
df['norm'] = df['norm'].apply(lambda x: word_tokenize(x))

# ‡∏ô‡∏±‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏≥
all_words = [word for tokens in df['norm'] for word in tokens]

# ‡∏ô‡∏±‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏≥
word_freq = Counter(all_words)

# ‡πÅ‡∏™‡∏î‡∏á‡∏Ñ‡∏≥‡∏ó‡∏µ‡πà‡∏û‡∏ö‡∏ö‡πà‡∏≠‡∏¢‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
print(word_freq.most_common(20))
```

# TF-IDF
#‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏≥

```python
from collections import Counter
import pandas as pd
import math

# ‡πÇ‡∏Ñ‡πâ‡∏î‡∏ô‡∏µ‡πâ‡∏î‡∏≥‡πÄ‡∏ô‡∏¥‡∏ô‡∏Å‡∏≤‡∏£‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡πÇ‡∏Ñ‡∏£‡∏á‡∏£‡πà‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô‡∏•‡∏¥‡∏™‡∏ï‡πå‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏≥‡πÉ‡∏ô‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå 'norm' ‡πÉ‡∏´‡πâ‡∏Å‡∏•‡∏≤‡∏¢‡πÄ‡∏õ‡πá‡∏ô‡∏•‡∏¥‡∏™‡∏ï‡πå‡∏Ñ‡∏≥‡πÄ‡∏î‡∏µ‡∏¢‡∏ß.
flat_tokens = [token for tokens in df['norm'] for token in tokens]

# ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏≥
tf_counter = Counter(flat_tokens)

# ‡∏ô‡∏±‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏≠‡∏Å‡∏™‡∏≤‡∏£‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡πÉ‡∏ô df
total_documents = len(df)

# ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡πà‡∏≤ IDF 
idf_values = {}
for term, tf in tf_counter.items():
    idf = math.log10(total_documents / (1 + tf))  # Using base 10 logarithm for IDF
    idf_values[term] = idf

# ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡πà‡∏≤ IDF ‡∏•‡∏á‡πÉ‡∏ô df
tf_df = pd.DataFrame(tf_counter.items(), columns=['Word', 'TF'])
tf_df['IDF'] = tf_df['Word'].map(idf_values)

# ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡πà‡∏≤ TF-IDF ‡πÇ‡∏î‡∏¢‡∏Ñ‡∏π‡∏ì‡∏Ñ‡πà‡∏≤ TF ‡πÅ‡∏•‡∏∞‡∏Ñ‡πà‡∏≤ IDF 
tf_df['TF-IDF'] = tf_df['TF'] * tf_df['IDF']

# ‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏•‡∏≥‡∏î‡∏±‡∏ö‡∏Ñ‡πà‡∏≤ TF-IDF ‡πÉ‡∏ô DataFrame tf_df ‡∏à‡∏≤‡∏Å‡∏°‡∏≤‡∏Å‡πÑ‡∏õ‡∏´‡∏≤‡∏ô‡πâ‡∏≠‡∏¢
tf_df = tf_df.sort_values(by='TF-IDF', ascending=False)

# Display the DataFrame with TF, IDF, and TF-IDF
tf_df
```

# Post tagger

```python
from pythainlp.tag import pos_tag

# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏ó‡∏≥ POS Tagging
def pos_tagging(tokens):
    tags = pos_tag(tokens)
    return tags

# ‡πÉ‡∏ä‡πâ DataFrame ‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå "Tokens" ‡πÅ‡∏•‡∏∞‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÉ‡∏´‡∏°‡πà "POS_Tags"
df['post'] = df['norm'].apply(pos_tagging)
df['post']
```

# LDA

```python
# Import necessary libraries
from gensim import corpora, models

# Tokenized and stop words removed data from the 'stopword' column
texts = df['norm']

# Create a dictionary
dictionary = corpora.Dictionary(texts)

# Create a corpus (a bag of words representation)
corpus = [dictionary.doc2bow(text) for text in texts]

# Train the LDA model
lda_model = models.LdaModel(corpus, num_topics=10, id2word=dictionary, passes=15)

# Print the topics
topics = lda_model.print_topics(num_words=5)
for topic in topics:
    print(topic)
```
